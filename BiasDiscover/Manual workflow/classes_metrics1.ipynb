          Pred
          A   B   C
          ---------
Actual A | AA  AB  AC

Actual B | BA  BB  BC

Actual C | CA  CB  CC


Where AA represents the number of true predictions of class A, AB represents the number of instances that belong to class A but were predicted as class B, and so on.

    Metrics Calculation for Each Class:

For class A:

    True Positives (TP) for class A = AA
    False Positives (FP) for class A = AB + AC (sum of row A excluding AA)
    True Negatives (TN) for class A = sum of all values outside of column A and row A
    False Negatives (FN) for class A = BA + CA (sum of column A excluding AA)

* Accuracy:
Accuracy=TP+TNTP+TN+FP+FNAccuracy=TP+TN+FP+FNTP+TN​

* Precision (Positive Predictive Value):
Precision=TPTP+FPPrecision=TP+FPTP​

* Recall (True Positive Rate or Sensitivity):
Recall=TPTP+FNRecall=TP+FNTP​

* Specificity (True Negative Rate):
Specificity=TNTN+FPSpecificity=TN+FPTN​

* F1 Score:
F1 Score=2×Precision×RecallPrecision+RecallF1 Score=2×Precision+RecallPrecision×Recall​

```python
from torchvision import datasets, transforms
```

```python
def get_label_mappings(data_dir):
    # Use a dummy transform since we just need the class names and indices
    transform = transforms.Compose([transforms.ToTensor()])
    
    # Load the dataset to access the class_to_idx attribute
    dataset = datasets.ImageFolder(data_dir, transform=transform)
    
    # Return the mapping
    # Return the reversed mapping
    return {v: k for k, v in dataset.class_to_idx.items()}
```

```python
classes_mapping = get_label_mappings("Plant_leave_diseases_dataset_without_augmentation")
classes_mapping
```

```python
import pandas as pd
```

```python
PATH = "../Plants-ConvNets-ViT/scripts/from_scratch/confusion.csv"
```

```python
df = pd.read_csv(PATH)
df
```

```python
confusion_matrix = pd.crosstab(df['True'], df['Predicted'], rownames=['True'], colnames=['Predicted'], margins=True)

# Extract 'All' column and row, and drop them from the original matrix
all_row = confusion_matrix.loc['All', :]
all_column = confusion_matrix['All']
confusion_matrix = confusion_matrix.drop('All', axis=1).drop('All', axis=0)

# Now sort the confusion matrix columns
confusion_matrix = confusion_matrix.reindex(sorted(confusion_matrix.columns), axis=1)

# Append 'All' column and row back
confusion_matrix['All'] = all_column[:-1]  # Exclude the last value which is the grand total
confusion_matrix.loc['All', :] = all_row.values

confusion_matrix
```

```python
metrics_dict = {}

for cls in confusion_matrix.columns[:-1]:  # Excluding the 'All' column
    TP = confusion_matrix.loc[cls, cls]
    FN = confusion_matrix.loc[cls, 'All'] - TP
    FP = confusion_matrix.loc['All', cls] - TP   #  sum of the row - TP
    TN = confusion_matrix.loc['All', 'All'] - TP - FP - FN

    accuracy = (TP + TN) / (TP + TN + FP + FN)
    precision = TP / (TP + FP)
    recall = TP / (TP + FN)

    f1 = 2 * precision * recall / (precision + recall)
    
    metrics_dict[cls] = {
        'Accuracy' : accuracy,
        'Precision': precision,
        'Recall': recall,
        'F1 Score': f1
    }
```

```python
# To view the metrics dictionary
for class_name, metrics in metrics_dict.items():
    print(f"Class: {class_name}")
    for metric_name, metric_value in metrics.items():
        print(f"\t{metric_name}: {metric_value:.4f}")
    print()
```

```python
# Convert dictionary to DataFrame
df_metrics = pd.DataFrame(metrics_dict).transpose()
df_metrics
```

```python
df_metrics.index = df_metrics.index.to_series().replace(classes_mapping)
```
