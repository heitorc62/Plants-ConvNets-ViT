```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Sample data from the CSV
df = pd.read_csv("../labelling_tool/results/dataframes/batch_0004_results.csv")

df.head()
```

```python
def plot_with_labels(df, standout_x=None, standout_y=None):
    plt.figure(figsize=(10, 8))
    colors = plt.cm.rainbow(np.linspace(0, 1, len(df['pred'].unique())))

    for i, label in enumerate(sorted(df['pred'].unique())):
        subset = df[df['pred'] == label]
        plt.scatter(subset['x'], subset['y'], color=colors[i], label=f'Label {label}')
    
    # Check if standout coordinates are provided
    if standout_x is not None and standout_y is not None:
        plt.scatter(standout_x, standout_y, color='yellow', edgecolor='black', linewidth=2, s=100, label='Standout Point')
    
    plt.legend()
    plt.xlabel('x')
    plt.ylabel('y')
    plt.title('Scatter plot of predicted labels')
    plt.show()
```

```python
plot_with_labels(df)
```

```python
aux_df = df[df["custom_data"] == 2947]
aux_df
```

```python
# Coordinates of the standout point
x_point = aux_df.iloc[0]["x"]
y_point = aux_df.iloc[0]["y"]

print(x_point, y_point)
plot_with_labels(df, x_point, y_point)
```
