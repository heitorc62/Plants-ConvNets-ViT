```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
from utils import get_label_mappings

DATASET_PATH = "/home/heitor/USP/IC/FAPESP/code_dataset/dataset/Plant_leave_diseases_dataset_without_augmentation"
```

```python
px_df = pd.read_csv("../results_8px/confusion.csv")

# Calculate confusion matrix
labels = np.unique(px_df[['True', 'Predicted']])
cm = confusion_matrix(px_df['True'], px_df['Predicted'], labels=labels)

# Plotting the confusion matrix
plt.figure(figsize=(25, 25))
sns.heatmap(cm, annot=True, fmt='d', xticklabels=labels, yticklabels=labels)
plt.title('Confusion Matrix for the 8px random forest')
plt.ylabel('Actual')
plt.xlabel('Predicted')
plt.show()
```

```python
labels_mapping = get_label_mappings(DATASET_PATH)
```

```python
regular_df = pd.read_csv("../results/regular_dataset/from_scratch/confusion.csv")
regular_df['True'] = regular_df['True'].map(labels_mapping)
regular_df['Predicted'] = regular_df['Predicted'].map(labels_mapping)
labels = np.unique(regular_df[['True', 'Predicted']])
regular_cm = confusion_matrix(regular_df['True'], regular_df['Predicted'], labels=labels)

# Plotting the confusion matrix
plt.figure(figsize=(25, 25))
sns.heatmap(regular_cm, annot=True, fmt='d', xticklabels=labels, yticklabels=labels)
plt.title('Confusion Matrix for the Regular Dataset')
plt.ylabel('Actual')
plt.xlabel('Predicted')
plt.show()
```

```python
y_true = np.array(regular_df["True"])
y_pred = np.array(regular_df["Predicted"])
print(classification_report(y_true, y_pred))
```
