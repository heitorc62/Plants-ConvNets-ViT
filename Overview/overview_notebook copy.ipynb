# **Problema de classificação:**
## **Dataset**:
Estamos estudando o dataset PlantVillage em um problema de classficação. Esse dataset é fruto do trabalho dos pesquisadores: David. P. Hughes e Marcel Salathe em artigo intulado "An open access repository of images on plant health to enable the development of mobile disease diagnostics" de novembro de 2015. Esse dataset é amplamente utilizado em problemas de classficação envolvendo Visão Computacional aplicada ao domínio da agricultura. Esse dataset é constituído de 5548 imagens divididas em 39 classes. 

Ainda, temos uma versão com data augmentation desse mesmo dataset. Contudo, tal versão não é contemplada nesse estudo.
